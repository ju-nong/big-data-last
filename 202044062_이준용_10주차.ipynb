2. class 별로 승객 수를 나타내는 countplot을 구하시오. (생존 여부('alive)'를 세부 분류로 함.)

```python
import seaborn as sns

titanic = sns.load_dataset('titanic')
sns.countplot(data=titanic,x='class', hue='alive')
```

3. class 별 count를 이용하여 pie chart를 작성 하시오.(image 참조)
a-3.png

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('titanic')

result=df.groupby('class')['class'].count()

plt.pie(result.values, labels=result.keys())
plt.show()
```

5. '결혼 유무'를 기준으로 하고 '종교 유무'로 하위 세부 분류 하는  빈도(countplot)를 나타내는 그래프를 **작성하시오**a-7.png

6.결혼 유무에 따른 파이 차트를 구하시오

```python
import matplotlib.pyplot as plt
import seaborn as sns

result=df.groupby('결혼 유무')['결혼 유무'].count()
plt.pie(result.values, labels=result.keys())
plt.show()
```

#국가별 입국자수 시각화

```python
import pandas as pd
df=pd.read_csv('/content/입국자.csv')
print(df)
```

9.(선택) image를 참조 하여 같은 그래프를 만들어 내는 code를 작성 하시오,b-9.png

```python
import seaborn as sns

ax = sns.lineplot(data=df, x='month', y='미국')
sns.lineplot(data=df, x='month', y='중국')
sns.lineplot(data=df, x='month', y='일본')
ax.set(xlabel='month', ylabel='입국자', title='국가별 입국자 수')
```

10. (선택) image를 참조 하여 같은  그래프를 만들어 내는 code를 작성 하시오, b-10.png

```python
import pandas as pd
import matplotlib.pyplot as plt

ay = df.plot(kind='barh', stacked=True, x='month', y=['미국', '중국', '일본'])
ay.set(title='월별 입국자 현황', xlabel='입국자 수', ylabel='month')
```
